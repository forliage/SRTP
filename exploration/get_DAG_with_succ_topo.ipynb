```python
import os
import json
import shutil
import re

# 源目录和目标目录
source_dir = r"D:\data\task_DAG_batch1"
target_dir = r"D:\data\DAG_with_succ_topo_batch1"

# 确保目标目录存在
os.makedirs(target_dir, exist_ok=True)

# 遍历源目录中的所有文件
for filename in os.listdir(source_dir):
    # 检查文件名是否为数字
    if filename.endswith('.json') and filename[:-5].isdigit():
        file_path = os.path.join(source_dir, filename)
        
        try:
            # 读取json文件
            with open(file_path, 'r', encoding='utf-8') as f:
                data = json.load(f)
                
            # 检查successful_topo字段是否存在且不为空
            if 'successful_topo' in data and data['successful_topo']:
                # 复制文件到目标目录
                shutil.copy2(file_path, os.path.join(target_dir, filename))
                print(f"已复制文件: {filename}")
                
        except json.JSONDecodeError:
            print(f"无法解析JSON文件: {filename}")
        except Exception as e:
            print(f"处理文件 {filename} 时发生错误: {str(e)}")

print("处理完成！")
```
