```python
from ant_api.utils import (
    get_default_config,
    ask_chatgpt
)
import base64
import os
import mimetypes
from typing import Optional

def encode_image_from_path(image_path: str, mime_type: Optional[str] = None) -> str:
    """
    Encode an image file to base64 string.
    :param image_path: The path of the image file.
    :param mime_type: The mime type of the image.
    :return: The base64 string.
    """

    file_name = os.path.basename(image_path)
    mime_type = (
        mime_type if mime_type is not None else mimetypes.guess_type(file_name)[0]
    )
    with open(image_path, "rb") as image_file:
        encoded_image = base64.b64encode(image_file.read()).decode("ascii")

    if mime_type is None or not mime_type.startswith("image/"):
        print(
            "Warning: mime_type is not specified or not an image mime type. Defaulting to png."
        )
        mime_type = "image/png"

    image_url = f"data:{mime_type};base64," + encoded_image
    return image_url

prompt_message = [
    {
        "role": "system", "content": [{"type": "text", "text": "You are a helpful assistant."}]
    },
    {
        "role": "user", "content": [
            {"type": "image_url", "image_url": {"url": encode_image_from_path(r"C:\Users\brindle\Desktop\screenshot_2024-12-24_22-13-35.png")}},
            {"type": "text", "text": "图中是什么东西？"}
        ]
    },
]
param = get_default_config(model="gpt-4o")
param["queryConditions"]["model"] = "gpt-4o"
param["queryConditions"]["temperature"] = "0.6"
param["queryConditions"]["messages"] = prompt_message
response = ask_chatgpt(param)
# print(param)
print(response)
```

```python
import json
import os
# 检查一下交叉验证过滤了多少subtask
path = "D:/data/logs_cross_validation"
# 先看下第一部分原始subtask数量
apps1 = ["Cursor","Zotero","Evernote","ShareX","Windows PowerShell ISE"]
tasks1 = []
for i in apps1:
    subtask_path = f"D:/data/logs_new_exploration/{i}_subtasks.json"
    with open(subtask_path, 'r') as f:
        task = json.load(f)
        tasks1 += task
print(f"第一部分原始subtask数量: {len(tasks1)}")
# print(tasks1[0])
# check一下过滤掉了多少subtask
cnt1 = 0
for i in tasks1:
    id = i["id"]
    for j in range(5):
        if os.path.exists(os.path.join(path, f"{id}({j})")):
            eval_res = json.load(open(os.path.join(path, f"{id}({j})", "eval_history.txt"), 'r'))
            if "success" in eval_res[-1] or "discriminative" in eval_res[-1]:
                # 在path目录下新建一个passed文件夹
                passed_dir = os.path.join(path, "passed")
                os.makedirs(passed_dir, exist_ok=True)
                # 将原目录复制到passed文件夹中
                src_dir = os.path.join(path, f"{id}({j})")
                dst_dir = os.path.join(passed_dir, f"{id}({j})")
                if not os.path.exists(dst_dir):
                    import shutil
                    try:
                        if os.path.exists(src_dir):
                            shutil.copytree(src_dir, dst_dir, ignore=shutil.ignore_patterns('Ak*'))
                    except Exception as e:
                        print(f"Error copying directory {src_dir}: {e}")
                cnt1 += 1
                break
        else:
            break
print(f"第一部分通过交叉验证数量: {cnt1}")
# 再看下第二部分原始subtask数量
apps2 = ["Microsoft Store","Mail","Calendar","People","Power Automate"]
tasks2 = []
for i in apps2:
    subtask_path = f"D:/data/logs_new_exploration/{i}_subtasks.json"
    with open(subtask_path, 'r') as f:
        task = json.load(f)
        tasks2 += task
print(f"第二部分原始subtask数量: {len(tasks2)}")
# print(tasks1[0])
# check一下过滤掉了多少subtask
cnt2 = 0
for i in tasks2:
    id = i["id"]
    for j in range(5):
        if os.path.exists(os.path.join(path, f"{id}({j})")):
            eval_res = json.load(open(os.path.join(path, f"{id}({j})", "eval_history.txt"), 'r'))
            if "success" in eval_res[-1] or "discriminative" in eval_res[-1]:
                # 在path目录下新建一个passed文件夹
                passed_dir = os.path.join(path, "passed")
                os.makedirs(passed_dir, exist_ok=True)
                # 将原目录复制到passed文件夹中
                src_dir = os.path.join(path, f"{id}({j})")
                dst_dir = os.path.join(passed_dir, f"{id}({j})")
                if not os.path.exists(dst_dir):
                    import shutil
                    try:
                        if os.path.exists(src_dir):
                            shutil.copytree(src_dir, dst_dir, ignore=shutil.ignore_patterns('Ak*'))
                    except Exception as e:
                        print(f"Error copying directory {src_dir}: {e}")
                cnt2 += 1
                break
        else:
            break
print(f"第二部分通过交叉验证数量: {cnt2}")
```
